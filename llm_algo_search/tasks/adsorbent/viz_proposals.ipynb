```python
import pickle
import textwrap

from ase.visualize import view
from ase.visualize.plot import plot_atoms
import matplotlib.pyplot as plt
```

```python
proposal_path = 'proposal_history_adsorbent.pkl'

with open(proposal_path, 'rb') as infile:
    proposals = pickle.load(infile)

for proposal in proposals:
    impl_cls = proposal.get_implementation()
    impl = impl_cls()
    try:
        adsorbent = impl.get_adsorbent()
    except:
        continue
    thought = proposal.get_thought()
    thought = '\n'.join(textwrap.wrap(thought, replace_whitespace=False, width=100))
    print(thought)
    plot_atoms(adsorbent)
    plt.show()
```

```python
proposal = proposals[-1]
impl_cls = proposal.get_implementation()
impl = impl_cls()
try:
    adsorbent = impl.get_adsorbent()
except Exception as e:
    print('Issue with implementation', e)
else:
    view(adsorbent)
```
